```python
import pandas as pd

df = pd.read_csv('data/Fish.csv')
df
```

```python
bream_df = df.loc[df['Species'] == 'Bream']

bream_length = bream_df['Length2'].tolist() # tolist ->같은 레벨에 있는 데이터끼리 묶어줌.
bream_weight = bream_df['Weight'].tolist()
```

```python
bream_length
```

```python
import matplotlib.pyplot as plt

# plot: 선 그래프, bar: 막대 그래프, scatter: 산점도 그래프
plt.scatter(bream_length, bream_weight)
```

```python
smelt_df = df.loc[df['Species'] == 'Smelt']
#smelt_df.head()

smelt_length = smelt_df['Length2'].tolist()
smelt_weight = smelt_df['Weight'].tolist()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
list(zip(length, weight))
```

```python
fish_data = [[1, w] for l, w in zip(length, weight)]  ## 튜플을 리스트로 바꾼 것. 꼭 []로 묶지 않아도 됌.
```

```python
# 정답 데이터 작성
fish_target = [1] * len(bream_weight) + [0] * len(smelt_weight)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
kn.predict([[30, 600]])
```

```python
kn.predict([[10, 20]])
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```
