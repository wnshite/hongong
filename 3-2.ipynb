```python
# 선형회귀 : 내가 갖고 잇는 데이터를 (명확하게 나타낼 수 있는) 가로지를 수 있는 가상의 선을 그리는 것. -- 수치 예상.
```

```python
import pandas as pd

df = pd.read_csv('data/Fish.csv')
df.head()
```

```python
perch_df = df.loc[df['Species'] == 'Perch']

perch_length = perch_df[['Length2']]
perch_weight = perch_df[['Weight']]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(perch_length, perch_weight)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()   # 인스턴스화 시키기

lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.predict([[100]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input, train_target)
plt.plot([15, 50], [15 * 36 - 647, 50 * 37 - 647])   # 선 그래프
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```
